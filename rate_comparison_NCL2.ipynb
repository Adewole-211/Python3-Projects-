```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df  =  pd.read_csv("flow_controller_headcount_defect.csv")
df.head(20)
```

```python
df.dtypes
```

```python
df_needed  =  df[['shift','day','process_path','interval_start_local','interval_end_local','dow','week','year_number_sun','predictedrate','actual_rate','plannedrate','rate_defect_type']]
```

```python
df_needed.head()
```

```python
df_needed['process_path'].value_counts()
```

```python
df_needed = df_needed[df_needed['process_path'].isin(['PPAFE1INDUCT', 'PPSINGLEMEDIUM', 'PPAFE1PACK', 'PPPICKTOREBIN2'])]
```

```python
df_needed['day'] = df_needed['day'].str.rstrip("T00:00:00.000Z")

df_needed.head()
```

```python
df_needed['interval_start_local'] = df_needed['interval_start_local'].str.replace('T',' ')
df_needed['interval_start_local'] = df_needed['interval_start_local'].str.replace('Z','')
df_needed['interval_end_local'] = df_needed['interval_end_local'].str.replace('T',' ')
df_needed['interval_end_local'] = df_needed['interval_end_local'].str.replace('Z','')


df_needed.head(10)
```

```python
df_needed['interval_end_local'] = pd.to_datetime(df_needed['interval_end_local'])
df_needed['interval_start_local'] = pd.to_datetime(df_needed['interval_start_local'])
df_needed['day'] = pd.to_datetime(df_needed['day'])

df_needed.dtypes
```

```python
df_needed = df_needed.sort_values(by = ['shift','interval_start_local','interval_end_local', 'dow'])
```

```python
df_needed['day'] = df_needed['interval_start_local'].dt.date
df_needed['interval_start_local'] = df_needed['interval_start_local'].dt.time
df_needed['interval_end_local'] = df_needed['interval_end_local'].dt.time

df_needed.head(10)

```

```python
df_needed = df_needed.drop(columns="year_number_sun")
```

```python
df_needed['shift'] = df_needed['shift'].str.strip()
```

```python
df_needed['shift'].str.len()
```

```python
df_ds = df_needed[df_needed['shift'] == 'Day']
df_ns = df_needed[df_needed['shift'] == 'Night']
```

```python
df_ds
```

```python
df_ns
```

```python
def assign_shift(dow):
    if hasattr(dow, 'strftime'):
        dow = dow.strftime('%A')
    day = dow.strip().title()  
    
    front_end = ['Sun', 'Mon', 'Tue']
    back_end = ['Thu', 'Fri', 'Sat']

    if day in front_end:
        return'Front End'
    elif day in back_end:
        return 'Back End'
    elif day == 'Wed':
        return 'Double Shift'
    else:
        return 'unknown'
```

```python
df_ds['fe_be'] = df_ds['dow'].apply(assign_shift)
df_ns['fe_be'] = df_ns['dow'].apply(assign_shift)
```

```python
df_ns
```

```python
df_ds
```

```python
df_needed['fe_be'] = df_needed['dow'].apply(assign_shift)
```

```python
df_needed.head(50)
```

```python
df_cleaned = df_needed[df_needed['plannedrate'] != 'null']
```

```python
cleaned_ds = df_ds[df_ds['plannedrate'] != 'null']
cleaned_ns = df_ns[df_ns['plannedrate'] != 'null']
```

```python
cleaned_ds
```

```python
cleaned_ns
```

```python
df_summary = df_cleaned.groupby(['process_path','fe_be', 'shift','interval_start_local', 'interval_end_local']).agg({'predictedrate':'mean','actual_rate': 'mean','plannedrate':'mean'}).reset_index().round(3)
df_summary.head(40)
```

```python
df_summary['Shift'] = (df_summary['fe_be'] + ' ' + df_summary['shift'])
df_summary.head(10)

```

```python
df_summary = df_summary.drop(columns = ['fe_be', 'shift'])
```

```python
df_summary.head(150)
```

```python
df_summary['interval_start_local'] = df_summary['interval_start_local'].apply(
    lambda x: x.strftime('%H:%M GMT') if pd.notna(x) else None
)
```

```python
df_summary['interval_end_local'] = df_summary['interval_end_local'].apply(
    lambda x: x.strftime('%H:%M GMT') if pd.notna(x) else None
)
```

```python
df_summary
```

```python
AFEINDUCT_summary  =  df_summary[df_summary['process_path'] == 'PPAFE1INDUCT']
```

```python
AFEPACK_summay  =  df_summary[df_summary['process_path'] == 'PPAFE1PACK']
PPPICKTOREBIN2_summary = df_summary[df_summary['process_path'] == 'PPPICKTOREBIN2']
PPSINGLEMEDIUM_summary = df_summary[df_summary['process_path'] == 'PPSINGLEMEDIUM']
```

```python
AFEINDUCT_summary = AFEINDUCT_summary.sort_values(by='interval_start_local')
```

```python
AFE_INDUCTTIME = AFEINDUCT_summary.groupby('interval_start_local').agg({'predictedrate':'mean','actual_rate': 'mean','plannedrate':'mean'}).reset_index().round(3)
```

```python


df_long = AFE_INDUCTTIME.melt(
    id_vars='interval_start_local',
    value_vars=['predictedrate', 'actual_rate', 'plannedrate'],
    var_name='rate_type',
    value_name='rate'
)

fig = px.line(
    df_long,
    x='interval_start_local',
    y='rate',
    color='rate_type',
    markers=True,
    text='rate',
    title='Average Rate Deviation by Time For AFE Induct',
    labels={ 'interval_start_local': 'Average Time Across Shifts',
            'rate': 'Aggregated Rates Across Shift',
            'rate_type': 'Rates Type'}
)
fig.update_traces(textposition='top center')
fig.show()
```

```python
sm_time = PPSINGLEMEDIUM_summary.groupby('interval_start_local').agg({'predictedrate':'mean','actual_rate': 'mean','plannedrate':'mean'}).reset_index().round(3)
```

```python
p2r2_time = PPPICKTOREBIN2_summary.groupby('interval_start_local').agg({'predictedrate':'mean','actual_rate': 'mean','plannedrate':'mean'}).reset_index().round(3)
afe_pack = AFEPACK_summay.groupby('interval_start_local').agg({'predictedrate':'mean','actual_rate': 'mean','plannedrate':'mean'}).reset_index().round(3)
```

```python
p2r2_long = p2r2_time.melt(
    id_vars='interval_start_local',
    value_vars=['predictedrate', 'actual_rate', 'plannedrate'],
    var_name='rate_type',
    value_name='rate'
)

fig = px.line(
    p2r2_long,
    x='interval_start_local',
    y='rate',
    color='rate_type',
    markers=True,
    text='rate',
    title='Average Rate Deviation by Time For Pick To Rebin 2 (PACK)',
    labels={ 'interval_start_local': 'Average Time Across Shifts',
            'rates': 'Aggregated Rates Across Shift',
            'rate_type': 'Rates Type'}
)
fig.update_traces(textposition='top center')
fig.show()
```

```python
sm_long = sm_time.melt(
    id_vars='interval_start_local',
    value_vars=['predictedrate', 'actual_rate', 'plannedrate'],
    var_name='rate_type',
    value_name='rate'
)

fig = px.line(
    sm_long,
    x='interval_start_local',
    y='rate',
    color='rate_type',
    markers=True,
    text='rate',
    title='Average Rate Deviation by Time For Single Medium',
    labels={ 'interval_start_local': 'Average Time Across Shifts',
            'rates': 'Aggregated Rates Across Shift',
            'rate_type': 'Rates Type'}
)
fig.update_traces(textposition='top center')
fig.show()
```
