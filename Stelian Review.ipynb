```python
brs1_clean['Id'] = pd.to_datetime(brs1_clean['Id'])
brs1_clean.info()
```

```python
brs1_clean['date_only'] = pd.to_datetime(brs1_clean['Id']).dt.strftime('%Y/%m/%d')
brs1_clean['time_only'] = pd.to_datetime(brs1_clean['Id']).dt.strftime('%H:%M')

brs1_clean.head(250)
```

```python
def assign_quarter(time_str):
    # Extract time part
    time_only = time_str.split() if ' ' in time_str else time_str
    hour, minute = map(int, time_only.split(':'))

    # Create time ranges as tuples (start_hour, start_min, end_hour, end_min)
    if (hour == 8 and minute >= 0) or (8 < hour < 10) or (hour == 10 and minute < 35):
        return 'Q1'
    elif (hour == 11 and minute >= 0) or (11 < hour < 14) or (hour == 14 and minute < 35):
        return 'Q2'
    elif (hour == 15 and minute >= 0) or (15 < hour < 18) or (hour == 18 and minute < 35):
        return 'Q3'
    elif (hour == 18 and minute >= 35) or (18 < hour < 20) or (hour == 20 and minute < 5):
        return 'Q1'
    elif (hour == 20 and minute >= 30) or (20 < hour < 24) or (0 <= hour < 1) or (hour == 1 and minute < 50):
        return 'Q2'
    elif (hour == 2 and minute >= 15) or (2 < hour < 5) or (hour == 5 and minute < 5):
        return 'Q3'
    else:
        return 'null'

brs1_clean['Quarter'] = brs1_clean['time_only'].apply(assign_quarter)
```

```python

def assign_shift(time_str):
    # Extract time part from "HH:MM GMT" format
    time_only = time_str.split() if ' ' in time_str else time_str
    hour, minute = map(int, time_only.split(':'))

    # Convert to decimal hours for proper comparison
    decimal_time = hour + minute / 60

    # Day Shift: 08:00 to 18:30 (8.0 to 18.5)
    if 8.0 <= decimal_time < 18.5:
        return 'DS'
    # Night Shift: 19:00 to 05:00 (wraps around midnight)
    elif decimal_time >= 19.0 or decimal_time < 5.0:
        return 'NS'
    else:
        return 'null'  # Gap between 18:30 and 19:00

brs1_clean['Shift'] = brs1_clean['time_only'].apply(assign_shift)
```

```python
# Convert the columns to numeric type (handles any non-numeric values)
brs1_clean['Label=ALPS Capacity [Current: {last}]'] = pd.to_numeric(
    brs1_clean['Label=ALPS Capacity [Current: {last}]'],
    errors='coerce'
)
brs1_clean['Label=Total OB Pack [last: {last}]'] = pd.to_numeric(
    brs1_clean['Label=Total OB Pack [last: {last}]'],
    errors='coerce'
)

# Remove rows where Quarter, Shift, or either numeric column has null/NaN values
brs1_clean_filtered = brs1_clean.dropna(subset=[
    'Quarter',
    'Shift',
    'Label=ALPS Capacity [Current: {last}]',
    'Label=Total OB Pack [last: {last}]'
])

# Also exclude rows where Quarter or Shift equals 'null' (as string)
brs1_clean_filtered = brs1_clean_filtered[
    (brs1_clean_filtered['Quarter'] != 'null') &
    (brs1_clean_filtered['Shift'] != 'null')
]

# Now group by Quarter and Shift, calculate averages
summary_table = brs1_clean_filtered.groupby(['Quarter', 'Shift']).agg({
    'Label=ALPS Capacity [Current: {last}]': 'mean',
    'Label=Total OB Pack [last: {last}]': 'mean'
}).reset_index()

# Rename columns for clarity
summary_table.columns = ['Quarter', 'Shift', 'Lagrange Value', 'OB PACK value']

# Round to 2 decimal places
summary_table['Lagrange Value'] = summary_table['Lagrange Value'].round(2)
summary_table['OB PACK value'] = summary_table['OB PACK value'].round(2)

# Calculate % Accuracy (percentage deviation) based on the averaged values
summary_table['% Accuracy'] = (
    (summary_table['OB PACK value'] - summary_table['Lagrange Value'])
    / summary_table['Lagrange Value']
) * 100
summary_table['% Accuracy'] = summary_table['% Accuracy'].round(2)

# Determine Status based on % Accuracy
summary_table['Status'] = np.where(
    abs(summary_table['% Accuracy']) <= 5,
    'Aligned',
    'Not Aligned'
)

# Reorder columns to match your specification
summary_table = summary_table[['Quarter', 'Shift', '% Accuracy', 'Lagrange Value', 'OB PACK value', 'Status']]

# Sort by Quarter and Shift
summary_table = summary_table.sort_values(by=['Quarter', 'Shift'])

summary_table.head()
```
